```python
left = pd.DataFrame({'key': ['foo', 'foo'], 'lval': [1, 2]})
right = pd.DataFrame({'key': ['foo', 'foo'], 'rval': [4, 5]})
print(left)
print(right)
```

# Reshaping

- stack :데이터프레임 열들의 계층을 “압축”합니다.

```python
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                     'foo', 'foo', 'qux', 'qux'],
                    ['one', 'two', 'one', 'two',
                     'one', 'two', 'one', 'two']]))

index = pd.MultiIndex.from_tuples(tuples,name=["first","second"])
df = pd.DataFrame(np.random.randn(8,2), index = index, columns=["A","B"])
#df
df2 = df[:4]
df2
```

```python
stacked = df2.stack()
stacked
```

- “Stack된” 데이터프레임 또는 (MultiIndex를 인덱스로 사용하는) Series인 경우, stack()의 역 연산은 unstack()이며, 기본적으로 마지막 계층을 unstack합니다.

```python
stacked.unstack()
print(stacked.unstack(1))
print(stacked.unstack(0))
```

# 범주화

```python
df = pd.DataFrame(
    {"id": [1,2,3,4,5,6],"raw_grade":["a","b","b","a","a","e"]}
)
df
```

```python
# 로우 데이터를 카테고리컬 데이터 타입으로 변환
df["grade"]=df["raw_grade"].astype("category")
df["grade"]
print(type(df["grade"]))
```

```python
# rename
df["grade"].cat.categories = ["very good","good","very bad"]
# 범주의 순서를 바꾸고 동시에 누락된 범주를 추가합니다 (Series.cat에 속하는 메소드는 기본적으로 새로운 Series를 반환합니다
df["grade"] = df["grade"].cat.set_categories(
    ["very bad","bad","medium","good","very good"]
)
df["grade"]
```

```python
df.sort_values(by="grade")
```

```python
df.groupby("grade").size()
```
